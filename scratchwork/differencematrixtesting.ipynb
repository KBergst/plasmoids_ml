```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = 3*np.sin(np.exp(np.linspace(-3,3,120)))
```

```python
# difference matrix
# A_ij = x_i *(1- abs(x_i - x_j)/range(x))

A = np.zeros((len(x), len(x)))
denom = np.ptp(x)
for i,x_i in enumerate(x):
    for j,x_j in enumerate(x):
        A[i,j] =x_i*(1 - np.abs(x_i - x_j)/denom)
```

```python
plt.imshow(A, cmap='binary')
plt.colorbar()
plt.show()
```

```python
plt.plot(x)
plt.show()
```

```python
# difference-squared matrix
# A_ij = x_i *(1- (x_i - x_j)^2/range(x)^2)

A = np.zeros((len(x), len(x)))
denom = np.ptp(x)
for i,x_i in enumerate(x):
    for j,x_j in enumerate(x):
        A[i,j] =x_i*(1 - (x_i - x_j)**2/denom**2)
```

```python
plt.imshow(A, cmap='binary')
plt.colorbar()
plt.show()
```
