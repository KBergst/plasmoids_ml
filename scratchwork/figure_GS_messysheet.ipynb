```python
import h5py
import matplotlib.pyplot as plt
```

### Open the desired h5 file and pull out the data we're interested in plotting

```python
fname = '/tigress/kendrab/analysis-notebooks/grad_shaf_outputs/perturbation_set_bigvar/perturbed_harris_20230320_110942.h5'

with h5py.File(fname, 'r') as file:
    print(file.keys())
    x_mgrid = file['x_mgrid'][:]
    z_mgrid = file['z_mgrid'][:]
    ay = file['ay'][:]
    jy = file['jy'][:]
    n = file['n'][:]
    
```

### Plot them all nice

```python
%matplotlib widget
fig, ax = plt.subplots(3, figsize=(7,7))

contour_ay = ax[0].contourf(z_mgrid, x_mgrid, ay, levels=25)
fig.colorbar(contour_ay, ax=ax[0])
ax[0].set(title=r"a) Flux function $A_{y}$", xlabel=r"z $\left(d_{i}\right)$", ylabel=r"x $\left(d_{i}\right)$")

contour_jy = ax[1].contourf(z_mgrid, x_mgrid, jy, levels=25)
fig.colorbar(contour_jy, ax=ax[1])
ax[1].set(title=r"b) Current component $j_{y}$", xlabel=r"z $\left(d_{i}\right)$", ylabel=r"x $\left(d_{i}\right)$")

contour_n = ax[2].contourf(z_mgrid, x_mgrid, n, levels=25)
fig.colorbar(contour_n, ax=ax[2])
ax[2].set(title=r"c) Plasma density $n$", xlabel=r"z $\left(d_{i}\right)$", ylabel=r"x $\left(d_{i}\right)$")

fig.tight_layout()
fig.savefig("/tigress/kendrab/analysis-notebooks/grad_shaf_outputs/testing/perturbed_IC_figure.svg")
plt.show()
```
