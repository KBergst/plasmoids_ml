```python
import sys
sys.path.append('/tigress/kendrab/python_pkgs')

import numpy as np
import math
import pyspedas
import pytplot

import mms_structure_finder.curlometer_basic as cb
```

```python
# get the data and interpolate the correct distances
for i in range(1,5):
    pyspedas.mms.fgm(probe=str(i), trange=['2019-09-27/12-47-00', '2019-09-27/18-55-30'], data_rate='brst',
                     time_clip=True)
    pyspedas.mms.mec(probe=str(i), trange=['2019-09-27/12-47-00', '2019-09-27/18-55-30'], data_rate='brst',
                     time_clip=True)
    # interpolate the mec data to the b data
    pyspedas.tinterpol(f'mms{i}_mec_r_gse', f'mms{i}_fgm_b_gse_brst_l2', method='linear',
                       newname=f'mms{i}_fgm_r_gse_brst_l2')
```

```python
# check for nans
for i in range(1,5):
    print(pytplot.data_quants[f'mms{i}_fgm_b_gse_brst_l2'][0:50])
    print(pytplot.data_quants[f'mms{i}_mec_r_gse'])
```

```python
#pyspedas version
pyspedas.mms.curlometer(fields=[f'mms{i}_fgm_b_gse_brst_l2' for i in range(1,5)],
                        positions=[f'mms{i}_mec_r_gse' for i in range(1,5)])
```

```python
# sync to common timeline
for i in range(1,5):
    pytplot.data_quants[f'mms{i}_fgm_b_gse_brst_l2'] = \
        pytplot.data_quants[f'mms{i}_fgm_b_gse_brst_l2'].interp(method="linear", assume_sorted=False,
                                                                time=pytplot.data_quants["mms1_fgm_b_gse_brst_l2"].time)
    pytplot.data_quants[f'mms{i}_fgm_r_gse_brst_l2'] = \
        pytplot.data_quants[f'mms{i}_fgm_r_gse_brst_l2'].interp(method="linear", assume_sorted=False,
                                                                time=pytplot.data_quants["mms1_fgm_b_gse_brst_l2"].time)
```

```python
# check shapes
for i in range(1,5):
    print(pytplot.data_quants[f'mms{i}_fgm_b_gse_brst_l2'].shape)
    print(pytplot.data_quants[f'mms{i}_fgm_r_gse_brst_l2'].shape)
```

```python
# package everything up
m=[]

for i in range(1,5):
    m_i = [pytplot.data_quants[f"mms{i}_fgm_b_gse_brst_l2"][:,j] for j in range(0,3)] + [pytplot.data_quants[f"mms{i}_fgm_r_gse_brst_l2"][:,j] for j in range(0,3)]
    print(len(m_i[0]))
    m.append(m_i)
```

```python
j, divb, divbcurlb = cb.curlometer_vec(m[0], m[1], m[2], m[3])
```
